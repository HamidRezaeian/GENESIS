## Experiment 30: The Falsification Test

Three-way ablation (200k ticks per arm, identical seeds):

| Arm | Condition | Population | Accuracy |
|-----|-----------|:----------:|:--------:|
| A | STDP ON | 373 | **43.3%** |
| B | STDP OFF (NOLEARN) | 600 | 2.9% |
| C | COSTONLY (weights frozen) | 600 | **56.9%** |

**Verdict:** STDP is actively harmful. Frozen weights predict 13.6 pp BETTER
than STDP-modified weights. Maladaptive weight drift is the root cause.

## The Fix (Deployed)

1. **Homeostatic STDP** (λ=0.01): `w += Δw_STDP − λ(w − w_DNA)` — anchors weights to DNA baseline
2. **Compositional Memory (CAM)**: 8-slot per-organism key-value store — substrate for working memory

Both 100% `@njit`-safe. Applied to `neuromorphic_engine.py` + `genesis_lab.py`.
